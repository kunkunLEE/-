```python
data_path = "D:/本科研究生资料/硕士研究生/课程/机器学习与数据科学/作业/cvpr20_id_dataset.txt"
def get_data_1(data_path):#cvpr16_id_dataset.txt
    with open(data_path,"r") as f:    #设置文件对象
        data = f.read() 
        data = data.split('\n')[2].split(' , ')
        data = sorted([int(value) for value in data])
    return data

def get_data_2(data_path):#cvpr19_id_dataset.txt
    with open(data_path,"r") as f:    #设置文件对象
        data = f.read() 
        data = data.split('\n')[1:]
        data = sorted([int(value) for value in data])
    return data

def get_data_3(data_path):#cvpr20_id_dataset.txt
    with open(data_path,"r") as f:    #设置文件对象
        data = f.read() 
        data = data.split('\n')[:-1]
        data = sorted([int(value) for value in data])
    return data
data = get_data_3(data_path)
print(data)
```

```python
import math
PI = math.pi
E = math.e

def f_Guass(x):
    y = (E**(-(x**2)/2))/((2*PI)**0.5)
    return y

def f_Laplace(x):
    y = (E**(-abs(x)))/2
    return y
    
def f_Epanechikov(x):
    if x<1 and x>-1:
        y = 0.75*(1 - (x**2))
    else:
        y = 0
    return y 

def f_Uniform(x):
    if x<1 and x>-1:
        y = 0.5
    else:
        y = 0
    return y 
```
