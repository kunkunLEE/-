```python
import numpy as np
import collections
from sklearn.model_selection import train_test_split
from scipy.io import loadmat
```

```python
d = loadmat('MNIST784x50x10.mat')
d.keys()
```

```python
x = d['X'].T
y = d['index'][0].reshape(500,1)
```

```python
result = 'k值,轮次,分类正确样本数,准确率,均方差\n'
```

```python
for k in [1, 20, 50]:
    acc_total = 0
    delta_total = 0
    for j in range(10):
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)
        test_num = len(x_test)
        y_predict = np.zeros(test_num)
        for i in range(test_num):
            distance = np.sum(np.square(x_train - x_test[i,:]), axis=1)
            nearest_idx = np.argpartition(distance, k-1)[0:k]
            votes = np.squeeze(y_train[nearest_idx])
            if k == 1:
                votes = np.array([votes])
            y_predict[i] = np.argmax(np.bincount(votes))
        y_test = np.squeeze(y_test)
        num = np.sum(y_predict==y_test)
        acc = np.sum(y_predict==y_test)/y_predict.shape[0]
        delta = (np.sum((y_predict-y_test)**2)/y_predict.shape[0])**(1/2)
        acc_total += acc
        delta_total += delta
        result = result + str(k) + ',' + str(j) + ',' + str(num) + ',' + str(acc) + ',' + str(delta) + '\n'
        print(k, j, num, acc, delta)
    result = result + str(k) + ',' + ',' +  ',' + str(acc_total/10) + ',' + str(delta_total/10) + '\n'
    print(k, acc_total/10, delta_total/10)

with open('knn_minist.csv', 'w') as f:
    f.write(result)
```
